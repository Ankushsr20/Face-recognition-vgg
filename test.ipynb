```python
from keras.models import load_model
```

```python
from keras.preprocessing import image
```

```python
his_img = image.load_img("sss/sappy_test2.jpg", target_size=(224, 224))
```

```python
his_img
```

```python
import numpy as np
```

```python
his_img = image.img_to_array(his_img)
```

```python
exp_his_img = np.expand_dims(his_img, axis=0)
```

```python
print(exp_his_img.shape)
```

```python
model = load_model('FCmodel123.h5')
```

```python
pred = model.predict(exp_his_img)
```

```python
if pred[0][1] == 1:
    print("Saptarsi")
else:
    print("Saranya")
```
